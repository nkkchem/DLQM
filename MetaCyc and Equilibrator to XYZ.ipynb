```python
import pandas as pd
import codecs
from IPython.display import HTML
```

# QM9 compounds

```python
qm9 = pd.read_table('../graph-neural-networks/mpnn-data/qm9/dsgdb9nsd.inchi',header=None,names=['cpd','InChI'])
qm9['cpd'] = qm9['cpd'].str.split(':').str.get(0)
```

```python
Cinchi = qm9[qm9['InChI'].str.contains(r'InChI=1S/C')]
Cinchi['nC'] = Cinchi['InChI'].str.extract(r'C(\d+)').fillna(1).astype(int)
qm9['nC'] = Cinchi['nC']
qm9['nC'] = qm9['nC'].fillna(0).astype(int)
qm9.sort_values('nC',ascending=False)
```

# MetaCyc compounds

```python
file_name = 'inchi-and-dG-of-metacyc.tab'
with codecs.open(file_name, "r",encoding='utf-8', errors='ignore') as fdata:
    metacyc = pd.read_table(fdata,index_col='FrameId')
metacyc
```

# MetaCyc compounds in QM9

```python
qm9_metacyc = metacyc.join(qm9.set_index('InChI'),on='InChI',how='inner')
print('''{} metacyc compounds
{} qm9 compounds
{} compounds in both qm9 and metacyc
{} unique metacyc compounds in overlap 
{} repeated metacyc compounds in overlap
{:0.2f}% of metacyc metabolites are in qm9 
{:0.2f}% of qm9 metabolites are in metacyc'''.\
      format(len(metacyc), 
             len(qm9), 
             len(qm9_metacyc),
             len(set(qm9_metacyc.index)),
             len(set(qm9_metacyc.loc[qm9_metacyc.index.duplicated()].index)),
             len(qm9_metacyc)/len(metacyc)*100,
             len(qm9_metacyc)/len(qm9)*100))
```
